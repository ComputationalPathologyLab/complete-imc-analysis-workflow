# Final Analysis Report

This notebook generates a reusable final interpretation report from the workflow outputs. The report is designed to update when the same pipeline is rerun on another compatible IMC dataset.

The report integrates:

- single-cell feature processing summary
- processed feature QC
- phenotype composition
- spatial interaction summary
- spatial enrichment
- plasma-cell-like niche analysis

The generated conclusion is exploratory and reflects the current rule-based phenotype workflow. It should not be treated as a final biological claim without validated annotation and statistical testing.


## Step 0: Configure Workflow Paths

This setup cell checks for the report-generation script and the key input tables needed to create the final interpretation summary.


````python
from pathlib import Path
import subprocess

current = Path.cwd().resolve()
if current.name == 'notebooks':
    WORKFLOW_DIR = current.parent
else:
    WORKFLOW_DIR = current

RESULTS_DIR = WORKFLOW_DIR / 'results'
SCRIPTS_DIR = WORKFLOW_DIR / 'scripts'
REPORT = RESULTS_DIR / '25_final_interpretation_summary.md'

required = [
    RESULTS_DIR / '11_processed_single_cell_features_summary.csv',
    RESULTS_DIR / '12_processed_single_cell_qc_summary.csv',
    RESULTS_DIR / '15_phenotype_composition_by_category.csv',
    RESULTS_DIR / '19_spatial_phenotype_interaction_summary.csv',
    RESULTS_DIR / '21_spatial_phenotype_enrichment_by_category.csv',
    RESULTS_DIR / '24_plasma_cell_niche_summary.csv',
]

print('Workflow directory:', WORKFLOW_DIR)
print('Report script exists:', (SCRIPTS_DIR / '14_generate_final_interpretation_report.py').exists())
for path in required:
    print(path.name, 'exists:', path.exists())
````

## Step 1: Generate Final Interpretation Summary

The report generator reads numbered workflow outputs and writes a Markdown report. The report is data-driven: category names, dominant phenotypes, enriched spatial pairs, depleted spatial pairs, and plasma-cell-like niche summaries are inferred from the current result tables.

A minimum edge-count threshold is applied when ranking enriched and depleted spatial pairs in the final narrative. This prevents rare one-off neighbor events from dominating the interpretation while preserving the complete enrichment tables for detailed review.

Output:

```text
results/25_final_interpretation_summary.md
```


````python
cmd = [
    'python3',
    str(SCRIPTS_DIR / '14_generate_final_interpretation_report.py'),
    '--workflow-dir', str(WORKFLOW_DIR),
    '--output', 'results/25_final_interpretation_summary.md',
]

result = subprocess.run(cmd, cwd=WORKFLOW_DIR, text=True, capture_output=True, check=True)
print(result.stdout)
````

## Step 2: Review Final Report Text

The generated Markdown report should be reviewed before use in slides, manuscripts, or project summaries. The language is intentionally cautious because the current annotation layer is rule-based and exploratory.
